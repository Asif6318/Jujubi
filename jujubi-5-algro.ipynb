```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms
import timm
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# Data preparation and split
def prepare_data(data_dir="/kaggle/input/jujubies/Jujubi leaf", batch_size=32, split_ratio=(0.7, 0.15, 0.15)):
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    # Load full dataset
    full_dataset = datasets.ImageFolder(data_dir, transform=transform)

    # Calculate split sizes
    total_size = len(full_dataset)
    train_size = int(split_ratio[0] * total_size)
    val_size = int(split_ratio[1] * total_size)
    test_size = total_size - train_size - val_size

    # Split dataset
    train_dataset, val_dataset, test_dataset = random_split(
        full_dataset, [train_size, val_size, test_size]
    )

    # Create data loaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    return train_loader, val_loader, test_loader, full_dataset.classes

# Your existing functions
def train_setup(m, lr=3e-4):
    return m, "cpu", nn.CrossEntropyLoss(), optim.Adam(m.parameters(), lr=lr)

def to_device(batch, device):
    return batch[0].to(device), batch[1].to(device)

# Enhanced metrics function
def get_metrics(model, ims, gts, loss_fn):
    preds = model(ims)
    loss = loss_fn(preds, gts)

    # Calculate multiple metrics
    pred_labels = torch.argmax(preds, dim=1)
    accuracy = (pred_labels == gts).float().mean()

    return loss, accuracy.item(), pred_labels.cpu().numpy(), gts.cpu().numpy()

# Training function with history tracking
def train_model(model_name, train_loader, val_loader, test_loader, num_classes, epochs=10):
    print(f"\n{'='*50}")
    print(f"Training {model_name}")
    print(f"{'='*50}")

    # Create model
    if model_name == "rexnet_150":
        model = timm.create_model("rexnet_150", pretrained=True, num_classes=num_classes)
    elif model_name == "resnet50":
        model = timm.create_model("resnet50", pretrained=True, num_classes=num_classes)
    elif model_name == "efficientnet_b0":
        model = timm.create_model("efficientnet_b0", pretrained=True, num_classes=num_classes)
    elif model_name == "mobilenetv3_large":
        model = timm.create_model("mobilenetv3_large_100", pretrained=True, num_classes=num_classes)

    # Setup training
    model, device, criterion, optimizer = train_setup(model)

    # History tracking
    history = {
        'train_loss': [], 'train_acc': [],
        'val_loss': [], 'val_acc': [],
        'test_metrics': {}
    }

    best_val_acc = 0

    for epoch in range(epochs):
        print(f"\nEpoch {epoch+1}/{epochs}")

        # Training phase
        model.train()
        train_loss, train_acc = 0, 0
        batches = 0

        for batch in tqdm(train_loader, desc="Training"):
            images, labels = to_device(batch, device)

            optimizer.zero_grad()
            loss, acc, _, _ = get_metrics(model, images, labels, criterion)

            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            train_acc += acc
            batches += 1

        avg_train_loss = train_loss / len(train_loader)
        avg_train_acc = train_acc / len(train_loader)

        # Validation phase
        model.eval()
        val_loss, val_acc = 0, 0
        all_val_preds, all_val_labels = [], []

        with torch.no_grad():
            for batch in tqdm(val_loader, desc="Validation"):
                images, labels = to_device(batch, device)
                loss, acc, preds, true_labels = get_metrics(model, images, labels, criterion)

                val_loss += loss.item()
                val_acc += acc
                all_val_preds.extend(preds)
                all_val_labels.extend(true_labels)

        avg_val_loss = val_loss / len(val_loader)
        avg_val_acc = val_acc / len(val_loader)

        # Store history
        history['train_loss'].append(avg_train_loss)
        history['train_acc'].append(avg_train_acc)
        history['val_loss'].append(avg_val_loss)
        history['val_acc'].append(avg_val_acc)

        print(f"Train Loss: {avg_train_loss:.4f}, Train Acc: {avg_train_acc:.4f}")
        print(f"Val Loss: {avg_val_loss:.4f}, Val Acc: {avg_val_acc:.4f}")

        # Save best model
        if avg_val_acc > best_val_acc:
            best_val_acc = avg_val_acc
            torch.save(model.state_dict(), f"best_{model_name}.pth")

    # Final test evaluation
    print(f"\nFinal Testing for {model_name}...")
    test_metrics = evaluate_model(model, test_loader, device, criterion)
    history['test_metrics'] = test_metrics

    return history, model

# Comprehensive evaluation function
def evaluate_model(model, test_loader, device, criterion):
    model.eval()
    all_preds, all_labels = [], []
    test_loss = 0

    with torch.no_grad():
        for batch in tqdm(test_loader, desc="Testing"):
            images, labels = to_device(batch, device)
            loss, _, preds, true_labels = get_metrics(model, images, labels, criterion)

            test_loss += loss.item()
            all_preds.extend(preds)
            all_labels.extend(true_labels)

    # Calculate all metrics
    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, average='weighted', zero_division=0)
    recall = recall_score(all_labels, all_preds, average='weighted', zero_division=0)
    f1 = f1_score(all_labels, all_preds, average='weighted', zero_division=0)

    metrics = {
        'loss': test_loss / len(test_loader),
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    }

    print(f"Test Metrics - Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, "
          f"Recall: {recall:.4f}, F1-Score: {f1:.4f}")

    return metrics

# Plot learning curves
def plot_learning_curves(histories):
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle('Model Learning Curves', fontsize=16)

    for model_name, history in histories.items():
        epochs = range(1, len(history['train_loss']) + 1)

        # Loss plot
        axes[0, 0].plot(epochs, history['train_loss'], label=f'{model_name} Train')
        axes[0, 0].plot(epochs, history['val_loss'], label=f'{model_name} Val', linestyle='--')
        axes[0, 0].set_title('Loss')
        axes[0, 0].set_xlabel('Epochs')
        axes[0, 0].set_ylabel('Loss')
        axes[0, 0].legend()

        # Accuracy plot
        axes[0, 1].plot(epochs, history['train_acc'], label=f'{model_name} Train')
        axes[0, 1].plot(epochs, history['val_acc'], label=f'{model_name} Val', linestyle='--')
        axes[0, 1].set_title('Accuracy')
        axes[0, 1].set_xlabel('Epochs')
        axes[0, 1].set_ylabel('Accuracy')
        axes[0, 1].legend()

    # Test metrics comparison
    model_names = list(histories.keys())
    test_accuracies = [histories[name]['test_metrics']['accuracy'] for name in model_names]
    test_f1_scores = [histories[name]['test_metrics']['f1_score'] for name in model_names]

    x = np.arange(len(model_names))
    width = 0.35

    axes[1, 0].bar(x - width/2, test_accuracies, width, label='Accuracy', alpha=0.7)
    axes[1, 0].bar(x + width/2, test_f1_scores, width, label='F1-Score', alpha=0.7)
    axes[1, 0].set_title('Test Performance Comparison')
    axes[1, 0].set_xlabel('Models')
    axes[1, 0].set_ylabel('Score')
    axes[1, 0].set_xticks(x)
    axes[1, 0].set_xticklabels(model_names, rotation=45)
    axes[1, 0].legend()

    # Detailed test metrics
    metrics_data = []
    for model_name in model_names:
        metrics = histories[model_name]['test_metrics']
        metrics_data.append([metrics['accuracy'], metrics['precision'], metrics['recall'], metrics['f1_score']])

    metrics_data = np.array(metrics_data)
    x = np.arange(len(model_names))
    width = 0.2

    for i, (metric_name, color) in enumerate(zip(['Accuracy', 'Precision', 'Recall', 'F1-Score'],
                                               ['blue', 'green', 'orange', 'red'])):
        axes[1, 1].bar(x + i*width, metrics_data[:, i], width, label=metric_name, alpha=0.7, color=color)

    axes[1, 1].set_title('Detailed Test Metrics')
    axes[1, 1].set_xlabel('Models')
    axes[1, 1].set_ylabel('Score')
    axes[1, 1].set_xticks(x + 1.5*width)
    axes[1, 1].set_xticklabels(model_names, rotation=45)
    axes[1, 1].legend()

    plt.tight_layout()
    plt.savefig('learning_curves.png', dpi=300, bbox_inches='tight')
    plt.show()

# Main function
def main():
    # Prepare data
    print("Loading and splitting data...")
    train_loader, val_loader, test_loader, classes = prepare_data()
    print(f"Classes: {classes}")
    print(f"Number of classes: {len(classes)}")

    # Define models to train
    models_to_train = ["rexnet_150", "resnet50", "efficientnet_b0", "mobilenetv3_large"]

    # Train all models
    histories = {}
    for model_name in models_to_train:
        history, _ = train_model(
            model_name, train_loader, val_loader, test_loader,
            num_classes=len(classes), epochs=10
        )
        histories[model_name] = history

    # Plot results
    plot_learning_curves(histories)

    # Print final comparison
    print("\n" + "="*60)
    print("FINAL MODEL COMPARISON")
    print("="*60)
    for model_name in models_to_train:
        metrics = histories[model_name]['test_metrics']
        print(f"\n{model_name}:")
        print(f"  Accuracy:  {metrics['accuracy']:.4f}")
        print(f"  Precision: {metrics['precision']:.4f}")
        print(f"  Recall:    {metrics['recall']:.4f}")
        print(f"  F1-Score:  {metrics['f1_score']:.4f}")



if __name__ == "__main__":


    # Uncomment below to train with real data
    main()
```
